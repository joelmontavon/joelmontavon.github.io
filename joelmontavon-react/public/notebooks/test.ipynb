## Healthcare Example: PDC Calculation

Proportion of Days Covered (PDC) is a adherence measure.

```python
# Simple PDC calculation example
days_covered = 312
measurement_period = 365

pdc = (days_covered / measurement_period) * 100
is_adherent = pdc >= 80

print(f"PDC: {pdc:.1f}%")
print(f"Adherent: {is_adherent} (threshold is 80%)")
```
